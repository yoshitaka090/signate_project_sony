```python
# カラムの上限、下限値を設定し、その間で都市ごとに平均値を求める
def trim_mean(City_all,train,train_columns,train_df):
  for i in City_all:
    City = train[train.City == i]
    one_list = []
    for col in train_columns:
      if col == 'temperature_min':
        tmean =  scipy.stats.tmean(City[col],(-30,30))
        one_list.append(tmean)
      elif col == 'temperature_mid':
        tmean = scipy.stats.tmean(City[col],(-20,40))
        one_list.append(tmean)
      elif col == 'temperature_max':
        tmean = scipy.stats.tmean(City[col],(-10,49))
        one_list.append(tmean)
      elif col == 'humidity_min':
        tmean = scipy.stats.tmean(City[col],(0,95))
        one_list.append(tmean)
      elif col == 'humidity_mid':
        tmean = scipy.stats.tmean(City[col],(5,95))
        one_list.append(tmean)
      elif col == 'humidity_max':
        tmean = scipy.stats.tmean(City[col],(5,99))
        one_list.append(tmean)
      elif col == 'pressure_min':
        tmean = scipy.stats.tmean(City[col],(880,1060))
        one_list.append(tmean)
      elif col == 'pressure_mid':
        tmean = scipy.stats.tmean(City[col],(880,1070))
        one_list.append(tmean)
      elif col == 'pressure_max':
        tmean = scipy.stats.tmean(City[col],(920,1080))
        one_list.append(tmean)
      elif col == 'ws_min':
        tmean = scipy.stats.tmean(City[col],(0,25))
        one_list.append(tmean)
      elif col == 'ws_mid':
        tmean = scipy.stats.tmean(City[col],(0,25))
        one_list.append(tmean)
      elif col == 'ws_max':
        tmean = scipy.stats.tmean(City[col],(1,30))
        one_list.append(tmean)
      else:
        tmean = scipy.stats.trim_mean(City[col],0.25)
        one_list.append(tmean)
    train_df.loc[i] = one_list
  return train_df
```

```python
tmean_train = trim_mean(City_train_all, train, train_columns, train_df)
tmean_test = trim_mean(City_test_all, test, test_columns, test_df)
```
